```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv")
display(df)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
x=df[['Experience Years']]
y=df['Salary']
plt.scatter(x,y)
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.figure(figsize=(10,3))
plt.show()
```

```python
df.shape
```

```python
x = df[['Experience Years']]
y = df['Salary']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
Year=float(input("enter you Experience in year:"))
salary=lr.predict([[Year]])
print(salary)
```

```python
x=df[['Experience Years']]
y=df['Salary']
plt.scatter(x,y)
plt.plot(x,lr.predict(x), color="pink")
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.figure(figsize=(10,3))
plt.show()
```
